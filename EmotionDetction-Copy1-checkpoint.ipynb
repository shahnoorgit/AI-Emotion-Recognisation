```python
import matplotlib.pyplot as plt#ploting
from glob import glob #finding files 
import cv2 #
import random #random Numbers
import os #interacting with os making dir
from tensorflow.keras.preprocessing.image import ImageDataGenerator #image Variations like (e.g., rotations, flips, color jittering) 
from tensorflow.keras.layers import Dense,Input,Dropout,Conv2D,Flatten #Dense: Fully connected layer for linear transformations.
#Input: Input layer for specifying the model's input shape.
#Dropout: Layer that randomly drops units during training to prevent overfitting.
#Conv2D: Convolutional layer for extracting features from images.
#Flatten: Layer that flattens the output of a convolutional layer into a vector.

from tensorflow.keras.layers import BatchNormalization,Activation,MaxPooling2D ##BatchNormalization: Layer that normalizes activations of the previous layer.
#Activation: Layer that applies activation functions (e.g., ReLU, sigmoid) to introduce non-linearity.
#MaxPooling2D: Layer that performs downsampling to reduce the dimensionality of feature maps.

from tensorflow.keras.models import Model,Sequential #stacking layers Sequentially

from tensorflow.keras.optimizers import Adam #The optimizer is responsible for updating the weights of the model during training based on the calculated loss

from tensorflow.keras.callbacks import ModelCheckpoint,ReduceLROnPlateau 
#Saves the model periodically during training based on certain criteria (e.g., best validation accuracy).
#Reduces Learning rate when perticular LR reaches low LR
from tensorflow.keras.utils import plot_model #ploting
import tensorflow as tf #This imports the core TensorFlow library as tf. TensorFlow is the underlying framework used for building and training deep
print(tf.__version__)
```

```python
img_size=48
batch_size=64
datagen_train= ImageDataGenerator()
train_generator= datagen_train.flow_from_directory("train/",
                                                   target_size=(img_size,img_size),
                                                   color_mode="grayscale",
                                                   batch_size=batch_size,
                                                   class_mode="categorical",
                                                   shuffle=True,
                                                  )

datagen_validation= ImageDataGenerator()
train_validation= datagen_train.flow_from_directory("test/",
                                                   target_size=(img_size,img_size),
                                                   color_mode="grayscale",
                                                   batch_size=batch_size,
                                                   class_mode="categorical",
                                                   shuffle=True,
                                                  )
```

Defining the model Layers

```python
def Convolution(input_tensor,filters,kernel_size):
    x=Conv2D(filters=filters,kernel_size=kernel_size,padding="same")(input_tensor)
    x=BatchNormalization()(x) #avoiding overfitting
    x=Activation('relu')(x)
    x=MaxPooling2D(pool_size=(2,2))(x)
    x=Dropout(0.25)(x)
    return x





#Convolution:

#Imagine applying a small filter (like a sliding window) to the image. This filter extracts specific features, like edges or corners, from the image data.
#The code uses multiple filters to capture different types of features.
#Padding ensures the output size remains the same, preserving spatial information.
#Batch Normalization:This step helps stabilize the training process and improve performance by normalizing the values of the previous layer's activations. Think of it like adjusting the brightness and contrast to make the features clearer.
#Activation:The ReLU activation function introduces non-linearity, allowing the model to learn more complex relationships between features. It essentially adds a threshold, making only positive values contribute to the next layer.
#Max Pooling:This reduces the dimensionality of the data by taking the maximum value within a specific region (e.g., 2x2 block). This helps make the model more robust to small shifts in the image and reduces computational cost.

#Dropout:This technique randomly drops a certain percentage of activations during training. It prevents the model from overfitting by forcing it to rely on a broader set of features, making it more generalizable.

#In simpler terms, this code block extracts features from the image, normalizes them, applies a non-linearity, reduces the data size, and introduces randomness to prevent overfitting. These steps are commonly used in CNNs to learn high-level representations from images for various tasks like classification.
```

```python
def Dense_f(input_tensor,nodes):
    x= Dense(nodes)(input_tensor)
    x= BatchNormalization()(x)
    x= Activation('relu')(x)
    x=Dropout(0.25)(x)
    return x

#x = Dense(nodes)(input_tensor): This line creates a dense layer using the Dense function (likely from a deep learning library like TensorFlow or Keras). The Dense function is commonly used to create fully connected layers in neural networks. Here's what happens:
#The nodes argument specifies the number of neurons in the dense layer.
#The input_tensor is passed as the input to this dense layer.
#Batch normalization is a technique used to improve the training stability and performance of neural networks by normalizing the activations of neurons across different mini-batches of data.
#The ReL Dropout randomly drops a certain percentage of activations (in this case, 25%) during training, helping to prevent the network from overfitting to the training data and improving its generalization ability.
#
```

```python
def model_fer(input_shape):
    inputs= Input(input_shape)
    conv_1= Convolution(inputs,32,(3,3))
    conv_2= Convolution(inputs,64,(5,5))
    conv_3= Convolution(inputs,128,(3,3))
    flatten=Flatten()(conv_3)
    dense_1 = Dense_f(flatten,256)
    output=Dense(7,activation="softmax")(dense_1)
    model= Model(inputs=[inputs],outputs=[output])
    model.compile(loss=['categorical_crossentropy'],optimizer='adam', metrics=['accuracy'])
    return model


#inputs = Input(input_shape): This line defines the input layer of the model, which takes data of a specific shape (input_shape). This shape likely represents the dimensions of the input images (e.g., height, width, and channels).
#conv_1 = Convolution(inputs, 32, (3, 3)): This creates a convolutional layer with 32 filters of size 3x3, applied to the input (inputs).
#These convolutional layers extract features from the input images, identifying patterns and edges relevant to emotion recognition.
#flatten = Flatten()(conv_3): This line flattens the output of the third convolutional layer (conv_3), converting the 3D tensor into a 1D vector. This is necessary before feeding the data into a fully connected layer.
#dense_1 = Dense_f(flatten, 256): This line calls the previously defined function Dense_f to create a dense layer with 256 neurons. The flattened output (flatten) from the previous step is passed as input to this dense layer. Recall that Dense_f likely performs the following:
#This line defines the output layer of the model. It creates a dense layer with 7 neurons
#model = Model(inputs=[inputs], outputs=[output]): This line constructs the overall CNN model by specifying the input and output layers.
#model.compile(loss=['categorical_crossentropy'], optimizer='adam', metrics=['accuracy']): This line compiles the model, defining the loss function (categorical crossentropy for multi-class classification), optimizer (Adam), and metrics (accuracy) to be used during training.
```

```python
model=model_fer((48,48,1))
model.summary()
#The argument (48, 48, 1) specifies the input shape for the model. This means the model expects input images that are 48 pixels high, 48 pixels wide, and have one color channel (likely grayscale).
```

Initailising the model

```python
epochs=15
steps_per_epoch= train_generator.n//train_generator.batch_size
validation_steps= train_validation.n//train_validation.batch_size

cheackpoint=ModelCheckpoint("model_weights.h5",monitor='val_accuracy',save_weights_only= True,mode='max',verbose=1)
callbacks=[cheackpoint]
#An epoch is a complete pass through the entire training dataset. Training the model for 15 epochs means it will iterate through the training data 15 times.
#steps_per_epoch= train_generator.n//train_generator.batch_size: This line calculates the number of steps (iterations) required per epoch
```

Training The Model

```python
model.fit(
x= train_generator,
steps_per_epoch= steps_per_epoch,
epochs=epochs,
validation_data=train_validation,
validation_steps= validation_steps,
callbacks= callbacks)

#callbacks ensures the checkpoint saves the model's weights when validation accuracy improves.
```

Model Evaluation


```python
model.evaluate(train_validation)
```
